```python
from pyspark.sql import functions as F
from pyspark.sql.types import ArrayType, StringType

# 1) Normalizacja tekstu: pojedyncze → podwójne cudzysłowy, żeby zrobić z tego poprawny JSON
cats_norm = F.when(
    (F.col("categories").isNull()) | (F.col("categories") == "") | (F.col("categories") == "NaN"),
    F.lit(None)
).otherwise(
    F.regexp_replace(F.col("categories"), "'", '"')
)

# 2) Parsowanie JSON → Array<String>
books_parsed = books_df.withColumn(
    "categories_arr",
    F.from_json(cats_norm, ArrayType(StringType()))
)

# (opcjonalny fallback) jeśli czasem masz wartości bez nawiasów, rozdziel po przecinku
# Użyj tylko gdy categories_arr pozostało null, ale w stringu jest przecinek.
books_parsed = books_parsed.withColumn(
    "categories_arr",
    F.when(
        F.col("categories_arr").isNull() & F.col("categories").rlike(","),
        F.split(F.col("categories"), r"\s*,\s*")
    ).otherwise(F.col("categories_arr"))
)

# 3) Rozbicie na wiersze (explode_outer zachowa też rekordy bez kategorii jako null)
books_exploded = books_parsed.withColumn("category", F.explode_outer("categories_arr"))

# 4) Policz rozkład gatunków
genre_distribution = (
    books_exploded
    .where(F.col("category").isNotNull() & (F.length(F.trim("category")) > 0))
    .groupBy(F.trim(F.col("category")).alias("category"))
    .agg(F.count("*").alias("book_count"))
    .orderBy(F.desc("book_count"))
)

genre_distribution.show(20, truncate=False)
```

```python
import matplotlib.pyplot as plt

pdf = genre_distribution.toPandas()

plt.figure(figsize=(10, 6))
plt.barh(pdf["category"][:15], pdf["book_count"][:15])
plt.xlabel("Number of Books")
plt.ylabel("Genre / Category")
plt.title("Distribution of Books by Genre")
plt.gca().invert_yaxis()
plt.show()
```

```python
from pyspark.sql import functions as F
from pyspark.sql.types import ArrayType, StringType
import matplotlib.pyplot as plt
import numpy as np

# ——— preprocess: lower, tylko litery/spacje, split, explode
clean = (
    reviews_df
    .select(F.col("review/text").alias("text"))
    .where(F.col("text").isNotNull() & (F.length(F.col("text")) > 0))
    .withColumn("text", F.lower("text"))
    .withColumn("text", F.regexp_replace("text", "[^a-z\\s]", " "))  # tylko litery
    .withColumn("tokens", F.split(F.regexp_replace("text", "\\s+", " ").cast("string"), " "))
)

# stopwords (możesz dodać własne)
stop = set("""
a an the and or of to in is it this that for with on as by from be are was were been being
i you he she we they me my our your their them his her its not no yes do does did doing done
but so because if while than then too very more most much many can could would should may might
have has had having at into over under out up down off about after before also just one two three
book books read review reviews author authors
""".split())

# explode + filtr
words = (
    clean
    .withColumn("word", F.explode("tokens"))
    .where(F.length("word") >= 3)                       # min. długość
    .where(~F.col("word").isin(list(stop)))             # usuń stopwords
    .groupBy("word").agg(F.count("*").alias("count"))
    .orderBy(F.desc("count"))
    .limit(20)
)

pdf_words = words.toPandas().sort_values("count", ascending=True)

# ——— wykres (coolwarm)
plt.figure(figsize=(9, 6))
colors = plt.cm.coolwarm(np.linspace(0, 1, len(pdf_words)))
plt.barh(pdf_words["word"], pdf_words["count"], color=colors)
plt.xlabel("Frequency")
plt.title("Most Frequent Words in Reviews (Top 20)")
plt.tight_layout()
plt.savefig("most_frequent_words.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from pyspark.sql import functions as F

most_reviewed = (
    reviews_df
    .where(F.col("Title").isNotNull() & (F.length("Title") > 0))
    .groupBy("Title")
    .agg(F.count("*").alias("num_reviews"))
    .orderBy(F.desc("num_reviews"))
    .limit(10)
)

pdf_books = most_reviewed.toPandas().sort_values("num_reviews", ascending=True)

plt.figure(figsize=(10, 6))
colors = plt.cm.coolwarm(np.linspace(0, 1, len(pdf_books)))
plt.barh(pdf_books["Title"], pdf_books["num_reviews"], color=colors)
plt.xlabel("Number of Reviews")
plt.title("Most Reviewed Books (Top 10)")
plt.tight_layout()
plt.savefig("most_reviewed_books.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
from pyspark.sql import functions as F
import matplotlib.pyplot as plt
import numpy as np

# --- Prosty preprocessing tytułów
titles_clean = (
    books_df
    .select(F.col("Title").alias("title"))
    .where(F.col("title").isNotNull() & (F.length(F.col("title")) > 0))
    .withColumn("t", F.lower("title"))
    .withColumn("t", F.regexp_replace("t", "[^a-z\\s]", " "))   # tylko litery/spacje
    .withColumn("t", F.regexp_replace("t", "\\s+", " "))        # 1+ spacji -> 1
    .withColumn("tokens", F.split(F.trim("t"), " "))
)

# --- Stopwords (możesz dodać/zmienić)
stop = set("""
a an the and or of to in is it this that for with on as by from be are was were been
i you he she we they my our your their his her not no yes do does did doing done
but so if while than then very more most can could would should may might
into over under out up down off about after before also one two three
book books guide history life love new
""".split())

words = (
    titles_clean
    .withColumn("word", F.explode("tokens"))
    .where(F.length("word") >= 3)
    .where(~F.col("word").isin(list(stop)))
    .groupBy("word").agg(F.count("*").alias("count"))
    .orderBy(F.desc("count"))
    .limit(20)
)

pdf_words = words.toPandas().sort_values("count", ascending=True)

plt.figure(figsize=(9,6))
colors = plt.cm.coolwarm(np.linspace(0, 1, len(pdf_words)))
plt.barh(pdf_words["word"], pdf_words["count"], color=colors)
plt.xlabel("Frequency")
plt.title("Most Common Words in Book Titles (Top 20)")
plt.tight_layout()
plt.savefig("common_title_words.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
from pyspark.sql import functions as F
import matplotlib.pyplot as plt
import numpy as np

# --- Wyciągnij 4-cyfrowy rok z kolumny 'publishedDate'
books_year = (
    books_df
    .withColumn("year_str", F.regexp_extract(F.col("publishedDate").cast("string"), r"^(\d{4})", 1))
    # zamiast cast używamy try_cast, żeby uniknąć błędu NumberFormatException
    .withColumn("year", F.expr("try_cast(year_str as int)"))
    .where(F.col("year").isNotNull() & (F.col("year") >= 1800) & (F.col("year") <= 2030))
    .groupBy("year")
    .agg(F.count("*").alias("book_count"))
    .orderBy("year")
)

# --- Konwersja do Pandas
pdf_year = books_year.toPandas()

# --- Wykres (line plot)
plt.figure(figsize=(10, 5))
plt.plot(pdf_year["year"], pdf_year["book_count"], color="steelblue")
plt.xlabel("Year")
plt.ylabel("Number of Books")
plt.title("Distribution of Books by Year")
plt.tight_layout()
plt.savefig("books_by_year.png", dpi=300, bbox_inches="tight")
plt.show()
```
